# 1. Catfacts

```python
# Import requests library
import requests

google = requests.get("https://developers.google.com")
print("Google:", google.status_code)

cats = response = requests.get("https://cat-fact.herokuapp.com/facts")
print("cats:", cats.status_code)
```

```python
#print(cats.text)
```

```python
import json

#response
results = cats.json()
#results
```

```python
import pandas as pd
pd.set_option("display.max_rows", 0)

data = pd.DataFrame(results)
data.head(10)
```

```python
data.shape
```

```python
pd.set_option('display.max_colwidth', None)
data
```
